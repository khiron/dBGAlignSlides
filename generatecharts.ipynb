```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(n, L):
    return L ** n

# Generate values for n
n = np.linspace(2, 10)

# Create the plot
plt.figure(figsize=(10, 6))

# Generate the graph for each L
for L in [1000]:
    plt.plot(n, f(n, L), label=f'L={L/1e6}mbp')

# Add a title
plt.title("Computation cost to align 1kbp sequences")

# Label the axes
plt.xlabel("Number of sequences")
plt.ylabel("Number of operations")

# Set the range of the axes
plt.xlim(1, 10)
# plt.yscale('log')

# Show the plot
plt.show()
```
